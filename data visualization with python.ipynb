```python
state = pd.read_csv('book.csv')
```

```python
state
```

```python
state[state.district == 'Kuching']
```

```python
#Compare the population growth of kuching and Bau districts
```

```python
kuching = state[state.district == 'Kuching']
```

```python
kuching
```

```python
sa = state[state.district == 'Sri Aman']
```

```python
sa
```

```python
plt.plot(kuching.year, kuching.population_total / 10**4)
plt.plot(sa.year, sa.population_total / 10**4)
plt.legend(['Kuching', 'Sri Aman'])
plt.xlabel('year')
plt.ylabel('population')
plt.show()
```

```python
kuching.population_total
```

```python
kuching.population_total / kuching.population_total.iloc[0] * 100
```

```python
plt.plot(kuching.year, kuching.population_total / kuching.population_total.iloc[0] * 100)
plt.plot(sa.year, sa.population_total / sa.population_total.iloc[0] * 100)
plt.legend(['Kuching', 'Sri Aman'])
plt.xlabel('year')
plt.ylabel('population growth (first year = 100)')
plt.show()
```
